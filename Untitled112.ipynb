```python
bitcoin_data1 = bitcoin_data[['text']]
bitcoin_data1['date'] = bitcoin_data['date']
bitcoin_data1.columns = ['tweets',"date"]
```

```python
def cleanTwt(twt):
    twt = re.sub("#bitcoin", 'bitcoin', twt) # removes the '#' from bitcoin
    twt = re.sub("#Bitcoin", 'Bitcoin', twt) # removes the '#' from Bitcoin
    twt = re.sub('#[A-Za-z0-9]+', '', twt) # removes any string with a '#'
    twt = re.sub('\\n', '', twt) # removes the '\n' string
    twt = re.sub('https:\/\/\S+', '', twt) # removes any hyperlinks
    return twt
```

```python
bitcoin_data1['cleaned_tweets'] = bitcoin_data1['tweets'].apply(cleanTwt)
```

```python
bitcoin_data1.drop(0, inplace=True)
bitcoin_data1.drop('tweets', axis = 1, inplace=True)
bitcoin_data1.info()
```

```python
bitcoin_data1['date'] = pd.to_datetime(bitcoin_data1['date'])
bitcoin_data1.info()
```

```python

bitcoin_data1 = bitcoin_data1.sort_values(by='date', ascending=False)
bitcoin_data1["cleaned_tweets"] = bitcoin_data1.groupby(['date'])['cleaned_tweets'].transform(lambda x: ' '.join(x))
bitcoin_data1 = bitcoin_data1.drop_duplicates()
bitcoin_data1.reset_index(inplace = True, drop = True)
bitcoin_data1
```

```python
def getSubjectivity(twt):
    return TextBlob(twt).sentiment.subjectivity

# create a function to get the polarity
def getPolarity(twt):
    return TextBlob(twt).sentiment.polarity

# create two new columns called "Subjectivity" & "Polarity"
bitcoin_data1['subjectivity'] = bitcoin_data1['cleaned_tweets'].apply(getSubjectivity)
bitcoin_data1['polarity'] = bitcoin_data1['cleaned_tweets'].apply(getPolarity)
```

```python
bitcoin_data1.head()
```

```python
def getSentiment(score):
    if score < 0:
        return "negative"
    elif score == 0:
        return "neutral"
    else:
        return "positive"
```

```python
bitcoin_data1['sentiment'] = bitcoin_data1['polarity'].apply(getSentiment)

# show the data
bitcoin_data1.head()
```

#Testing the stationarity of a time series using the Dickey-Fuller test to capture shorter trends and changes

```python
def test_stationarity(timeseries):

 print('The following displays the Results of dickey fuller test')
 result = adfuller(timeseries, autolag = 'AIC')
 labels = ['ADF Test','p-value','#Lags Used','Number of Observations']
 for value,label in zip(result, labels):
   print(label+' : '+str(value) )
 if result[1] <= 0.05:
   print("Strong evidence against the null hypothesis, reject the null hypothesis.Which suggest that the Data is stationary")
 else:
   print("Weak evidence against null hypothesis,which suggest that the time series is non-stationary ")
test_stationarity(training_samples['Close'])
```
